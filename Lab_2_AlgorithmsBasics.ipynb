```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
# Exercise 2
def sum_arr(arr):
  total = 0
  for i in arr:
    total += i
  return total

# Sample function call
print(sum_arr([1, 2, 3, 4]))

# Time Complexity: O(n)
# Reason: The loop runs once for each element in the array (n elements).
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
# call the function
arr = [12, 3, 4, 15, 45, 78, 450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
# Exercise 3
def max_min(arr):
  minimum = arr[0]
  maximum = arr[0]

  for i in arr:
    if i < minimum:
      minimum = i
    if i > maximum:
      maximum = i

  return minimum, maximum

# Sample function call
arr = [12, 3, 4, 15, 45, 78, 450]
print(max_min(arr))

# Time Complexity: O(n)
# Reason: The function traverses the array once, comparing each element.
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
# Exercise 4
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count

# Sample function call
arr = [12, 3, 4, 15, 45, 78, 450]
print(Counting_EvenOdd(arr))

# Time Complexity: O(n)
# Reason: The loop goes through all n elements once, performing constant-time operations.
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
# Exercise 5
def linear_Search(arr, target):
    """
    Performs a linear search for the target element in the given array.
    Returns the index of the target if found, otherwise returns -1.
    """
    for index in range(len(arr)):
        if arr[index] == target:
            return index  # Element found
    return -1  # Element not found

# Sample function call
arr = [12, 3, 4, 15, 45, 78, 450]
target = 15
print(linear_Search(arr, target))

# Time Complexity: O(n)
# Reason: In the worst case, the function may have to check all n elements of the array.
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
# Exercise 6
def Binary_Search(arr, target):
    """
    Performs a binary search for the target element in a sorted array.
    Returns the index of the target if found, otherwise returns -1.
    Note: The array must be sorted.
    """
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  # Element found
        elif arr[mid] < target:
            left = mid + 1  # Search in the right half
        else:
            right = mid - 1  # Search in the left half

    return -1  # Element not found

# Sample function call
arr = [1, 3, 5, 7, 9]
target = 5
print(Binary_Search(arr, target))

# Time Complexity: O(log n)
# Reason: Each step reduces the search space by half.
```
